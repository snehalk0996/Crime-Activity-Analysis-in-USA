## **Data Exploration:**

The data exploration displays different features of the data given below:

a) Print head of data

b) Print column names

c) Print the number of rows and columns

d) Print type of columns

e) Sum of each column

```python
# Import necessary libraries

import pandas as pd
import numpy as np
```

```python
# Read dataset and print head

report_crime = pd.read_csv("report.csv")
print("First 10 rows of the dataset:")
report_crime.head(10)
```

```python
# Print column names

print("Columns Names:")
print(report_crime.columns)
```

```python
# Print number of rows and columns

num_rows, num_columns = report_crime.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Print the data types of each column

print(report_crime.dtypes)
```

```python
# Find sum of each column

total_crime = report_crime.sum()

# Print sum of each column

print(total_crime)
```

```python
# Import necessary library

import matplotlib.pyplot as plt

# Create variable for columns containing crime values

columns_new = ['violent_crimes', 'homicides', 'rapes', 'assaults', 'robberies',
       'months_reported', 'crimes_percapita', 'homicides_percapita',
       'rapes_percapita', 'assaults_percapita', 'robberies_percapita']

# Create a new dataframe with olumns containing crime values
new_df = report_crime[columns_new]

# Plot the data using Matplotlib
plt.figure(figsize=(12,6))
plt.bar(new_df.columns, new_df.mean(), color='midnightblue')
plt.xlabel('Columns')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.title('Bar Plot of Crime Count')
plt.show()
```

After plotting the above bar plot, we come to know that the crime rate of violent crimes including all regions is drastically high compared to other crimes.

We need to clean the data by finding and eliminating the missing values inorder to carry out further analysis.

```python
# Checking missing values

missing_values = report_crime.isnull().sum()

# Print missing values

print(missing_values)
```

```python
# Data Cleaning - removing missing values

new_crime_data = report_crime.dropna()

# Print 10 rows of clean data

new_crime_data.head(10)
```

# **One column selection:**

```python
# Find mode of column agency_jurisdiction

crime_city = new_crime_data['agency_jurisdiction'].mode().iloc[0]

# Print most crime occuring city

print(crime_city)
```

After finding out the most repeated number of city repeated in the dataset, we found that Albuquerue, NM is the city where large number of crime occurs.

```python
# Import counter class from collections library

from collections import Counter

# Find occurence of each city

count = Counter(new_crime_data['agency_jurisdiction'])

print(count)
```

We also found out the number of crimes happening in the cities of United States of America displaying from highest number of crimes to lowest number of crimes in a particular city.

```python
# Using the `nlargest` method to get the 10 largest values from the report_year column

report_year_records = new_crime_data.nlargest(10, 'report_year')

# Print 10 largest values of the report_year column

report_year_records.head(10)
```

We have selected report year column for further analysis. We used n.large() function to display the highest number in report years. Again, after arranging the data of report year in descending order, we could see Albuquerque, NM has the highest violent crime rate.

# **Sort:**

```python
# Create new variable with 5 columns for creating new dataset

new_columns = ['report_year', 'agency_jurisdiction', 'violent_crimes', 'crimes_percapita', 'homicides_percapita']

# Create new dataset order by 2 columns report_year and violent_crimes

new_ds = new_crime_data[new_columns].sort_values(by=['report_year', 'violent_crimes'], ascending=[False, False]).head(10)

# Print the sorted data

new_ds
```

Here we have created a new dataset with only 5 special columns: 'report_year', 'agency_jurisdiction', 'violent_crimes', 'crimes_percapita' and 'homicides_percapita'. We then sort this new dataset in descending order (highest to lowest) based on the report year and Violent Crime column using the sort_values() function. The head section (10) ensures that we will reach the 10 highest ranks of reort year and violent crime. This will give us an idea of which cities had the highest rates of violent crime in which year.

## **Series:**

```python
# Print information about 5 different columns

col_names = ['population', 'violent_crimes', 'assaults', 'robberies', 'crimes_percapita']

new_series = new_crime_data[col_names]

for column in col_names:

  print(f"Column: {column}")

  print(f"Mean: {new_series[column].mean()}")

  print(f"Median: {new_series[column].median()}")

  print(f"Standard Deviation: {new_series[column].std()}")

  print(f"Max Value: {new_series[column].max()}")

  print(f"Min Value: {new_series[column].min()}")

  print("/n")
```

In this section, we're creating a new dataFrame called new_series that contains information from 5 different columns: 'population', 'violent_crimes', 'assaults', 'robberies' and 'crimes_percapita'. Based on the bar plot of crime counts, we have selected these columns. These columns are selected on the basis of the count of crime rates. This dataframe is a subset of the original dataset. By doing this, we get brief information about these specific columns in our report. This allows us to highlight key data points and attributes that matter for our analysis and recommendations. This information shows mean, median, standard deviation, maximum and minimum values of that particular column.

## **Plot:**

```python
# Import necessary library

import matplotlib.pyplot as plt

# Create a bar plot for total violent crimes by report year

plt.figure(figsize=(19,6))
plt.bar(new_crime_data['report_year'], new_crime_data['violent_crimes'], label='violent_crimes', color='darkviolet')

# Add labels and title

plt.xlabel('Report Year')
plt.ylabel('Violent Crimes')
plt.title('Total Violent Crimes by Report Year')

# Show a legend to label the data

plt.legend()

# Display the plot

plt.show()
```

The above bar plot shows violent crimes by report year. X axis displays year from 1975 to 2015. Y axis shows count of violent crime. The year 1990 is the peak year where highest number of violent crimes happened. After 1990, we could the crime rate of violent crime decreasing till 2014. Even though the crime rate of violent crime is decreasing from 1990, still violent crime has the highest count among all other types of crime.
